```python
# Univariate Analysis - Analyze one column at a time like column by column
```

```python
predictors = ['BRId','GenderId','IAId','Amount of Credit Cards',
              'Nationality','Occupation','Fee Structure','Risk Weighting']

# Number of plots
num_plots = len(predictors)

# Set up the grid layout (e.g., 3 columns)
cols = 3
rows = (num_plots + cols - 1) // cols  # Ceiling division

# Set figure size (width, height)
plt.figure(figsize=(15, rows * 4))

for i, predictor in enumerate(predictors):
    plt.subplot(rows, cols, i + 1)
    sns.countplot(data=df, x=predictor)
    plt.title(predictor)
    plt.tight_layout()

plt.show()
```

```python
## Numerical Analysis
```

```python
numeric_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance',
                'Business Lending','Bank Loans','Bank Deposits',
                'Checking Accounts','Saving Accounts']

# Total number of plots
num_plots = len(numeric_cols)

# Grid size (e.g., 3 columns)
cols = 3
rows = (num_plots + cols - 1) // cols  # Ceiling division

# Set overall figure size
plt.figure(figsize=(15, rows * 4))

for i, col in enumerate(numeric_cols):
    plt.subplot(rows, cols, i + 1)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(col)
    plt.tight_layout()

plt.show()
```

```python
numeric_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance',
                'Business Lending','Bank Loans','Bank Deposits',
                'Checking Accounts','Saving Accounts']

# Calculate correlation matrix
corr = df[numeric_cols].corr()

# Set the figure size
plt.figure(figsize=(10, 8))

# Plot the heatmap with 'crest' colormap
sns.heatmap(corr, annot=True, cmap='crest', fmt=".2f", linewidths=0.5, linecolor='white', square=True)

plt.title("Correlation Heatmap of Numeric Columns", fontsize=14)
plt.tight_layout()
plt.show()
```

""" Interpreataion- 
Bank Deposits vs. Checking Accounts:

A high correlation suggests that when customers have higher checking account balances, they also tend to have higher total bank deposits.

Checking accounts might be a significant component of the total deposits.

Bank Deposits vs. Saving Accounts:

Similarly, customers with higher saving balances contribute more to overall bank deposits.
